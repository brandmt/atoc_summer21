```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Task 1: Calculate and compare heating rate profiles under clear-sky conditions for tropical, midlatitude summer, and sub-arctic summer atmospheres.

```python
cp = 1004.67 #J/kg/K

file = '/Users/brmo2880/Downloads/Tropical_Clear.csv'
data = pd.read_csv(file)

z = data.loc[:,'Height']
dFdz = data.loc[:,'dF/dz']
dTdt = -dFdz/cp

plt.figure(dpi=100)
plt.title('Clear-Sky Profile: Tropical')
plt.grid(alpha=0.2)
plt.plot(dTdt*86400,z/1000)
# plt.xscale('log')
plt.xlabel('Heating Rate (K/day)')
plt.ylabel('Height (km)')
```

## Task 2: Calculate and compare heating rate profiles with high or low clouds in the tropical atmosphere.

```python
file = '/Users/brmo2880/Downloads/Tropical_Clouds.csv'
data = pd.read_csv(file)

z = data.loc[:,'Height']
dFdz = data.loc[:,'dF/dz']
dTdt = dFdz/cp

plt.figure(dpi=100)
plt.title('High Clouds Profile: Tropical')
plt.grid(alpha=0.2)
plt.plot(dTdt*86400,z/1000)
plt.xscale('log')
plt.xlabel('Cooling Rate (K/day)')
plt.ylabel('Height (km)')
```
